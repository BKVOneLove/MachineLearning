Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.




```python
'''import csv

def qwer():
  pass

with open("/content/Global YouTube Statistics.csv", encoding='unicode_escape') as csvfile:
  spamreader'''

import csv
from collections import defaultdict

# Открываем файл CSV для чтения
with open('/content/Global YouTube Statistics.csv', mode='r', encoding='unicode_escape') as file:
    reader = csv.DictReader(file)

    # Создаем словарь для хранения данных по категориям
    category_data = defaultdict(list)

    for row in reader:
        # Замените 'category_column' на имя вашей колонки с категориальными данными
        category = row['Youtuber']
        # Замените 'numeric_column' на имя вашей числовой колонки
        numeric_value = float(row['subscribers'])

        # Добавляем числовое значение в соответствующую категорию
        category_data[category].append(numeric_value)

# Вычисляем медиану и среднее значение для каждой категории
result = {}
for category, values in category_data.items():
    median_value = sum(values) / len(values)  # Среднее значение
    average_value = sorted(values)[len(values) // 2]  # Медиана

    result[category] = {'Среднее значение': median_value, 'Медиана': average_value}

# Выводим результаты
for category, data in result.items():
    print(f'Категория: {category}')
    print(f'Среднее значение: {data["Среднее значение"]}')
    print(f'Медиана: {data["Медиана"]}')
    print()
```
