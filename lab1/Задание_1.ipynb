##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.




```python
import csv

category_name = 'channel_type'
non_number = ['Youtuber',
               'category',
               'Title',
               'Country',
               'Abbreviation',
               'channel_type',
               'created_year',
               'created_month',
               'created_date']

with open("/content/Global YouTube Statistics.csv", encoding='unicode_escape') as f:
    reader = csv.reader(f)
    headers = next(reader)
    aggregate_data = []
    cat_types = {}
    cat_index = int()
    for i in range(len(headers)):
      if headers[i] == category_name:
        cat_index = i
        break

    for row in reader:
        if len(row) == len(headers):
            for i in range(len(row)):
                try:
                    if headers[i] in non_number:
                        continue

                    value = float(row[i])

                    cat_type = row[cat_index]
                    category = headers[i]

                    if cat_type not in cat_types:
                        cat_types[cat_type] = len(cat_types)
                        aggregate_data.append({})



                    if category in aggregate_data[cat_types[cat_type]]:
                        aggregate_data[cat_types[cat_type]][category].append(value)
                    else:
                         aggregate_data[cat_types[cat_type]][category] = [value]


                except ValueError:
                    continue
        else:
            continue

for i in range(len(aggregate_data)):
    print('***************************')
    print(f'Channel_type: {list(cat_types.keys())[i]}')
    print('---')
    for category, values in aggregate_data[i].items():
        median = sorted(values)[len(values) // 2]
        mean = sum(values) / len(values)
        print(f'Category: {category}')
        print(f'Median: {median}')
        print(f'Mean: {mean}')
        print('---')
```
